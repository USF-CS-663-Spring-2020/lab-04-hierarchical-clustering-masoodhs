```
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

df = pd.read_csv('NIPS_1987-2015.csv')

df
```

Assign each word to it's related row (switch col[0] with rows)

```
df.rename(columns = {list(df)[0]: 'words'}, inplace = True)
df = df.set_index('words')
del df.index.name
df
```

Calculate the distance between each sample

```
Z = linkage(df, 'ward')
```

Plot the dendrogram.

```
plt.figure(figsize=(15, 10))
dendrogram(Z, truncate_mode = 'lastp', p=200 ) # -> 200 leaves at the bottom of the plot.
plt.show()
```
